```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train.head()
```

```python
df_train.shape
```

```python
from matplotlib import pyplot as plt
%matplotlib inline 
import seaborn as sns
sns.set_style('darkgrid')
```

```python
plt.figure(figsize= (14,6))
sns.boxplot(df_train)
plt.title("Boxplot-Graph")
```

```python
#Describe The Dataset
df_train.describe()
```

```python
train_X = df_train.drop(columns=['Item_Outlet_Sales','Outlet_Establishment_Year'],axis=1)
train_y = df_train['Item_Outlet_Sales']
```

```python
train_X.head()
```

```python
test_X = df_test.drop(columns=['Item_Outlet_Sales','Outlet_Establishment_Year'],axis=1)
test_y = df_test['Item_Outlet_Sales']
print(test_y)
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(train_X,train_y)
```

```python
preds_y = lr.predict(test_X)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mean_absolute_error(test_y,preds_y)
```
